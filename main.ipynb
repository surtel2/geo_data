```python

# 4. Поиск выбросов и аномалий через IQR (Outliers & Anomalies)
print("--- 3. ПОИСК СТАТИСТИЧЕСКИХ ВЫБРОСОВ (IQR METHOD) ---")
# Выбираем только числовые колонки для анализа аномалий
numeric_cols = df.select_dtypes(include=[np.number]).columns

outliers_summary = []
for col in numeric_cols:
    # Игнорируем ID-колонки и коды регионов, так как для них IQR не имеет смысла
    if col.lower() in [
        "objectid",
        "id",
        "stfips",
        "urban_code",
        "version",
        "dir",
    ] or "fips" in col.lower():
        continue

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Считаем количество строк за пределами усов
    outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[
        0
    ]
    outliers_pct = (outliers_count / len(df)) * 100

    outliers_summary.append(
        {
            "Column": col,
            "Min": df[col].min(),
            "Max": df[col].max(),
            "Outliers Count": outliers_count,
            "Outliers (%)": round(outliers_pct, 2),
        }
    )

outliers_df = pd.DataFrame(outliers_summary).sort_values(
    by="Outliers (%)", ascending=False
)
print(outliers_df[outliers_df['Outliers Count']>0])
print("\n" + "=" * 50 + "\n")
```

```python

# 5. Логические проверки на "Inaccurate / Inconsistent Data"
print("--- 4. ЛОГИЧЕСКИЕ ПРОТИВОРЕЧИЯ ---")

# Проверка 1: Отрицательные значения там, где их быть не должно (длина, скорость, объемы)
negative_checks = {}
for col in numeric_cols:
    neg_count = (df[col] < 0).sum()
    if neg_count > 0:
        negative_checks[col] = neg_count

if negative_checks:
    print("⚠️ Найдены отрицательные значения в строго положительных полях:")
    for k, v in negative_checks.items():
        print(f"   - Колонка '{k}': {v} строк с отрицательными значениями.")
else:
    print("✅ Тест на отрицательные значения пройден (все показатели >= 0).")

# Проверка 2: Аномальные ограничения скорости (например, Speed_Limit == 0 или > 100 миль/ч)
if "Speed_Limit" in df.columns:
    bad_speed = df[(df["Speed_Limit"] <= 0) | (df["Speed_Limit"] > 90)].shape[0]
    print(f"   - Строк с нереалистичным Speed_Limit (<=0 или >90): {bad_speed}")

# Проверка 3: Нулевая длина при наличии данных
if "LENGTH" in df.columns:
    zero_len = df[df["LENGTH"] <= 0].shape[0]
    print(f"   - Строк с нулевой или отрицательной длиной LENGTH: {zero_len}")
```

```python
import numpy as np
import pandas as pd

# Загрузка (если уже в памяти — этот шаг можно пропустить)
# df = pd.read_csv("FAF_NTD_2026_Data.csv")

print("--- ПОЛНЫЙ АУДИТ СТРУКТУРЫ И СИЛОСОВ ---")

# 1. Считаем точный процент пропусков для ВСЕХ 53 колонок (без ограничений топ-20)
total_rows = len(df)
all_missing = (
    df.isnull()
    .sum()
    .reset_index()
    .rename(columns={"index": "Column", 0: "Missing_Rows"})
)
all_missing["Missing_Percentage"] = (all_missing["Missing_Rows"] / total_rows) * 100
all_missing = all_missing.sort_values(by="Missing_Percentage", ascending=False)

print(f"Всего колонок в датасете: {len(all_missing)}")


# 2. Группируем колонки по твоим "кластерам силосов" для отчета
def cluster_sparsity(pct):
    if pct == 0:
        return "0% (Perfect/Core Fields)"
    elif pct <= 8:
        return "0% - 8% (Minor Incompleteness)"
    elif pct <= 33:
        return "8% - 33% (Regional/State Specific Data)"
    elif pct <= 55:
        return "51% - 52% (Specific Infrastructure Silo)"
    elif pct <= 70:
        return "68% (Niche Network Attributes)"
    elif pct <= 85:
        return "79% - 82% (High Sparsity / Federal Sub-networks)"
    elif pct <= 95:
        return "91% - 92% (Operational Adjustments Silo)"
    else:
        return "98% - 100% (Critical Sparsity / Edge Cases / Border Only)"


all_missing["Silo_Cluster"] = all_missing["Missing_Percentage"].apply(
    cluster_sparsity
)

print("\n📊 РАСПРЕДЕЛЕНИЕ КОЛОНОК ПО СИЛОСАМ (ДЛЯ РЕПОРТА):")
cluster_summary = (
    all_missing.groupby("Silo_Cluster")["Column"]
    .apply(list)
    .reset_index()
)
for idx, row in cluster_summary.iterrows():
    print(f"\n📁 Кластер: {row['Silo_Cluster']}")
    print(f"   Колонки ({len(row['Column'])} шт): {', '.join(row['Column'])}")

print("\n" + "=" * 50 + "\n")

# 3. ПРОВЕРКА SCHEMA DRIFT: Ломается ли схема в зависимости от VERSION?
print("--- АНАЛИЗ SCHEMA DRIFT ПО КОЛОНКЕ VERSION ---")
if "VERSION" in df.columns:
    versions = df["VERSION"].unique()
    print(f"Обнаруженные версии/периоды в данных 2026 года: {versions}")

    # Проверим, меняется ли заполненность ключевых полей от версии к версии
    for v in versions:
        v_df = df[df["VERSION"] == v]
        print(f"\nАнализ подмножества версии {v} (Строк: {len(v_df)}):")
        # Посмотрим топ-5 пустых колонок именно для этой версии
        v_missing = (v_df.isnull().sum() / len(v_df)) * 100
        print(v_missing[v_missing > 0].sort_values(ascending=False).head(5))
else:
    print("Колонка VERSION не найдена для анализа Drift.")

print("\n" + "=" * 50 + "\n")

# 4. МАТРИЦА КОРРЕЛЯЦИИ ПРОПУСКОВ (Показывает, какие колонки пустеют ОДНОВРЕМЕННО)
print("--- КОРРЕЛЯЦИЯ ПРОПУСКОВ (Идентификация системных Silos) ---")
# Отобрать только те колонки, где есть хоть какие-то пропуски
cols_with_nan = df.columns[df.isnull().any()].tolist()
if len(cols_with_nan) > 1:
    # Создаем булеву матрицу (True если Null)
    null_pattern = df[cols_with_nan].isnull()
    # Считаем корреляцию: если 1.0, значит колонки всегда пустые вместе (это один силос)
    null_corr = null_pattern.corr()

    # Выведем пары с идеальной корреляцией пропусков (1.0)
    high_corr_pairs = []
    for i in range(len(null_corr.columns)):
        for j in range(i + 1, len(null_corr.columns)):
            if null_corr.iloc[i, j] > 0.99:
                high_corr_pairs.append(
                    (null_corr.columns[i], null_corr.columns[j])
                )

    print(
        f"Найдено {len(high_corr_pairs)} пар колонок, которые пустеют ОДНОВРЕМЕННО (доказательство жестких Data Silos):"
    )
    for pair in high_corr_pairs[:15]:  # Выведем первые 15 пар
        print(f"   - {pair[0]} <---> {pair[1]}")
```

```python
import numpy as np
import pandas as pd

# 1. Создаем карту ожидаемых типов на основе скопированного тобой словаря
expected_schema = {
    # Core & IDs
    "ID": "Long integer",
    "OBJECTID": "Long integer",
    "LENGTH": "Double-precision decimal number",
    "DIR": "Long integer",
    "DATA1": "Long integer",
    "VERSION": "Character string",
    # Class & Route
    "Class": "Long integer",
    "Class_Description": "Character string",
    "Road_Name": "Character string",
    "Sign_Rte": "Character string",
    "Rte_Type": "Character string",
    "Rte_Number": "Character string",
    "Rte_Qualifier": "Character string",
    # Geography
    "Country": "Character string",
    "STATE": "Character string",
    "STFIPS": "Character string",  # Обрати внимание: в доке Character string!
    "County_Name": "Character string",
    "CTFIPS": "Character string",  # В доке Character string!
    "Urban_Code": "Character string",  # В доке Character string!
    "FAFZONE": "Long integer",
    # Infrastructure & Tolls
    "Status": "Long integer",
    "F_Class": "Long integer",
    "Facility_Type": "Long integer",
    "NHS": "Long integer",
    "STRAHNET": "Long integer",
    "NHFN": "Long integer",
    "Truck": "Character string",
    "AB_Lanes": "Long integer",
    "BA_Lanes": "Long integer",
    "Speed_Limit": "Long integer",
    "Toll_Type": "Long integer",
    "Toll_Name": "Character string",
    "Toll_Link": "Long integer",
    "Toll_Link_Name": "Character string",
    # HPMS & Borders
    "HPMS_USA_RouteID": "Character string",
    "HPMS_Begin_Point": "Double-precision decimal number",
    "HPMS_End_Point": "Double-precision decimal number",
    "BorderState1": "Character string",
    "BorderState2": "Character string",
    "BorderFAF1": "Long integer",
    "BorderFAF2": "Long integer",
    "TRUCKTOLL": "Double-precision decimal number",
    "BorderLink": "Long integer",
    "AddedBorderTime": "Double-precision decimal number",
    # Operational & Speeds
    "AdjustSpeed": "Double-precision decimal number",
    "AdjustReason": "Character string",
    "AB_FinalSpeed": "Double-precision decimal number",
    "BA_FinalSpeed": "Double-precision decimal number",
    "AB_CombinedSpeed": "Double-precision decimal number",
    "BA_CombinedSpeed": "Double-precision decimal number",
    "AB_FreeFlowTime": "Double-precision decimal number",
    "BA_FreeFlowTime": "Double-precision decimal number",
    "Shape__Length": "Double-precision decimal number",
}

print("--- АУДИТ СООТВЕТСТВИЯ ТИПОВ ДАННЫХ (DATA TYPE COMPLIANCE) ---")

discrepancies = []

for col, expected_type in expected_schema.items():
    if col not in df.columns:
        discrepancies.append(
            {"Column": col, "Expected": expected_type, "Actual": "MISSING COLUMN"}
        )
        continue

    actual_dtype = df[col].dtype
    null_count = df[col].isnull().sum()

    # Маппинг фактического типа на понятный язык
    if actual_dtype == "int64":
        actual_mapped = "Long integer"
    elif actual_dtype == "float64":
        # Если есть Null, то Long Integer превращается в float64
        if expected_type == "Long integer" and null_count > 0:
            actual_mapped = "Long integer (Broken to Float due to NaNs)"
        else:
            actual_mapped = "Double-precision decimal number"
    elif actual_dtype == "object":
        actual_mapped = "Character string"
    else:
        actual_mapped = str(actual_dtype)

    # Если типы не совпадают
    if (
        expected_type == "Long integer"
        and "Long integer" not in actual_mapped
    ) or (expected_type != "Long integer" and expected_type != actual_mapped):
        discrepancies.append(
            {
                "Column": col,
                "Expected": expected_type,
                "Actual": actual_mapped,
                "Null_Count": null_count,
            }
        )

# Вывод результатов
discrepancies_df = pd.DataFrame(discrepancies)
if not discrepancies_df.empty:
    print(
        f"⚠️ Найдено {len(discrepancies_df)} несоответствий между словарем и базой:"
    )
    print(discrepancies_df.to_string(index=False))
else:
    print("✅ Идеально! Все типы данных строго соответствуют официальной схеме.")
```
